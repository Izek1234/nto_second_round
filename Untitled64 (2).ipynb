```python
import zipfile
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
from catboost import CatBoostRegressor
from sentence_transformers import SentenceTransformer
import joblib
import os
from datetime import datetime

print("Загрузка данных из CSV...")
data_path = "/content"
train_df = pd.read_csv(f"{data_path}/train.csv")
test_df = pd.read_csv(f"{data_path}/test.csv")
books_df = pd.read_csv(f"{data_path}/books.csv")
users_df = pd.read_csv(f"{data_path}/users.csv")
genres_df = pd.read_csv(f"{data_path}/genres.csv")
book_genres_df = pd.read_csv(f"{data_path}/book_genres.csv")
book_descriptions_df = pd.read_csv(f"{data_path}/book_descriptions.csv")

print(f"Train shape: {train_df.shape}")
print(f"Test shape: {test_df.shape}")
print(f"Books shape: {books_df.shape}")
print(f"Users shape: {users_df.shape}")
```

```python
# --- Объединение данных ---
print("Объединение данных...")

# Объединение train с книгами и пользователями
train_full = train_df.merge(books_df, on='book_id', how='left')
train_full = train_full.merge(users_df, on='user_id', how='left')

# Объединение test с книгами и пользователями
test_full = test_df.merge(books_df, on='book_id', how='left')
test_full = test_full.merge(users_df, on='user_id', how='left')

# --- Добавление жанров ---
print("Добавление информации о жанрах...")

# Сначала объединяем book_genres_df с genres_df, чтобы получить названия жанров
book_genres_with_names = book_genres_df.merge(genres_df[['genre_id', 'genre_name']], on='genre_id', how='left')

# Группируем по book_id и собираем все названия жанров в список
book_genres_agg = book_genres_with_names.groupby('book_id')['genre_name'].apply(list).reset_index()
book_genres_agg.columns = ['book_id', 'genre_names']

# Присоединяем список жанров к train_full и test_full
train_full = train_full.merge(book_genres_agg, on='book_id', how='left')
test_full = test_full.merge(book_genres_agg, on='book_id', how='left')

# Заполняем пропуски (книги без жанров)
train_full['genre_names'] = train_full['genre_names'].fillna(pd.Series([[] for _ in range(len(train_full))], index=train_full.index))
test_full['genre_names'] = test_full['genre_names'].fillna(pd.Series([[] for _ in range(len(test_full))], index=test_full.index))

# --- Объединение с описаниями книг ---
train_full = train_full.merge(book_descriptions_df, on='book_id', how='left')
test_full = test_full.merge(book_descriptions_df, on='book_id', how='left')

# Заполнение пропусков в описаниях (для BERT)
train_full['description'] = train_full['description'].fillna("")
test_full['description'] = test_full['description'].fillna("")

print("Объединение данных завершено.")
```

```python


# --- Вычисление SBERT эмбеддингов ---
print("Загрузка модели SBERT...")
sbert_model_name = "ai-forever/ruBert-base"
sbert_model = SentenceTransformer(sbert_model_name)

# Получаем уникальные описания для вычисления эмбеддингов
all_descriptions_df = pd.concat([
    train_full[['book_id', 'description']].drop_duplicates(subset=['book_id']),
    test_full[['book_id', 'description']].drop_duplicates(subset=['book_id'])
]).drop_duplicates(subset=['book_id']).reset_index(drop=True)

print(f"Вычисление SBERT эмбеддингов для {len(all_descriptions_df)} уникальных описаний...")
descriptions_list = all_descriptions_df['description'].tolist()
sbert_embeddings = sbert_model.encode(descriptions_list, batch_size=64, show_progress_bar=True)

# Создание DataFrame с эмбеддингами
sbert_cols = [f'sbert_emb_{i}' for i in range(sbert_embeddings.shape[1])]
sbert_emb_df = pd.DataFrame(sbert_embeddings, columns=sbert_cols)
sbert_emb_df['book_id'] = all_descriptions_df['book_id'].values

print("SBERT эмбеддинги вычислены.")
```

```python


# --- Добавление SBERT эмбеддингов к основным датафреймам ---
print("Добавление SBERT эмбеддингов к датафреймам...")
train_full = train_full.merge(sbert_emb_df, on='book_id', how='left')
test_full = test_full.merge(sbert_emb_df, on='book_id', how='left')

# Заполняем пропуски (если были книги без описаний)
train_full[sbert_cols] = train_full[sbert_cols].fillna(0)
test_full[sbert_cols] = test_full[sbert_cols].fillna(0)

print("SBERT эмбеддинги добавлены.")

# --- Фильтрация обучающих данных (как в бейзлайне) ---
print("Фильтрация обучающих данных (has_read=1)...")
train_filtered = train_full[train_full['has_read'] == 1].copy()
# Убедимся, что timestamp в формате datetime
train_filtered['timestamp'] = pd.to_datetime(train_filtered['timestamp'])

# --- Временное разделение ---
print("Временное разделение обучающих данных...")
split_ratio = 0.8
train_filtered = train_filtered.sort_values('timestamp').reset_index(drop=True)
split_idx = int(len(train_filtered) * split_ratio)

train_part = train_filtered.iloc[:split_idx].copy()
val_part = train_filtered.iloc[split_idx:].copy()

print(f"Размер train_part: {train_part.shape[0]}")
print(f"Размер val_part: {val_part.shape[0]}")

```

```python

COL_USER_ID = 'user_id'
COL_BOOK_ID = 'book_id'
COL_AUTHOR_ID = 'author_id'
COL_RATING = 'rating'
COL_GENRE_ID = 'genre_id'
COL_DESCRIPTION = 'description'

# Ваши новые имена фичей
F_USER_MEAN_RATING = "user_mean_rating"
F_USER_RATINGS_COUNT = "user_ratings_count" # Обратите внимание на имя в бейзлайне
F_BOOK_MEAN_RATING = "book_mean_rating"
F_BOOK_RATINGS_COUNT = "book_ratings_count" # Обратите внимание на имя в бейзлайне
F_AUTHOR_MEAN_RATING = "author_mean_rating"
F_BOOK_GENRES_COUNT = "book_genres_count"

def create_aggregates(df, df_for_aggs=None):
    """Создаёт агрегаты по user_id, book_id, author_id."""
    if df_for_aggs is None:
        df_for_aggs = df

    # Агрегаты по пользователю (на основе df_for_aggs)
    user_agg = df_for_aggs.groupby(COL_USER_ID)[COL_RATING].agg(['mean', 'count', 'std']).reset_index()
    user_agg.columns = [COL_USER_ID, F_USER_MEAN_RATING, F_USER_RATINGS_COUNT, 'user_std_rating'] # Используем константу
    user_agg['user_std_rating'] = user_agg['user_std_rating'].fillna(0)

    # Агрегаты по книге (на основе df_for_aggs)
    book_agg = df_for_aggs.groupby(COL_BOOK_ID)[COL_RATING].agg(['mean', 'count', 'std']).reset_index()
    book_agg.columns = [COL_BOOK_ID, F_BOOK_MEAN_RATING, F_BOOK_RATINGS_COUNT, 'book_std_rating'] # Используем константу
    book_agg['book_std_rating'] = book_agg['book_std_rating'].fillna(0)

    # Агрегаты по автору (на основе df_for_aggs)
    author_agg = df_for_aggs.groupby(COL_AUTHOR_ID)[COL_RATING].agg(['mean', 'count', 'std']).reset_index()
    author_agg.columns = [COL_AUTHOR_ID, F_AUTHOR_MEAN_RATING, 'author_ratings_count', 'author_std_rating'] # Используем константу
    author_agg['author_std_rating'] = author_agg['author_std_rating'].fillna(0)

    # Присоединение агрегатов к df
    df = df.merge(user_agg, on=COL_USER_ID, how='left')
    df = df.merge(book_agg, on=COL_BOOK_ID, how='left')
    df = df.merge(author_agg, on=COL_AUTHOR_ID, how='left')

    # Заполнение пропусков для новых пользователей/книг/авторов
    # Используем общую среднюю оценку из df_for_aggs
    global_mean_rating = df_for_aggs[COL_RATING].mean()
    df[F_USER_MEAN_RATING] = df[F_USER_MEAN_RATING].fillna(global_mean_rating)
    df[F_USER_RATINGS_COUNT] = df[F_USER_RATINGS_COUNT].fillna(0)
    df['user_std_rating'] = df['user_std_rating'].fillna(0)

    df[F_BOOK_MEAN_RATING] = df[F_BOOK_MEAN_RATING].fillna(global_mean_rating)
    df[F_BOOK_RATINGS_COUNT] = df[F_BOOK_RATINGS_COUNT].fillna(0)
    df['book_std_rating'] = df['book_std_rating'].fillna(0)

    df[F_AUTHOR_MEAN_RATING] = df[F_AUTHOR_MEAN_RATING].fillna(global_mean_rating)
    df['author_ratings_count'] = df['author_ratings_count'].fillna(0)
    df['author_std_rating'] = df['author_std_rating'].fillna(0)

    return df

# --- Вычисление признака количества жанров (book_genres_count) ---
print("Вычисление признака количества жанров...")

# Функция для подсчёта длины списка в ячейке
def count_genres(genre_list):
    if isinstance(genre_list, list):
        return len(genre_list)
    else:
        # Если не список (например, NaN, который стал []), возвращаем 0
        return 0


print("Признак количества жанров добавлен.")

# ... (ваш предыдущий код, включая константы, create_aggregates, вычисление book_genres_count) ...

# --- Импорты для TF-IDF ---
from sklearn.feature_extraction.text import TfidfVectorizer
import joblib
import os

# --- Параметры TF-IDF ---
TFIDF_MAX_FEATURES = 500  # Максимальное количество признаков
TFIDF_MIN_DF = 2          # Минимальное количество документов для включения термина
TFIDF_MAX_DF = 0.95       # Максимальная доля документов для включения термина
TFIDF_NGRAM_RANGE = (1, 2) # Использовать униграммы и биграммы

# --- Вычисление TF-IDF признаков ---
print("Вычисление TF-IDF признаков...")

# Папка для сохранения модели векторайзера (опционально)
MODEL_DIR = "/content/models_tfidf" # Укажите путь для сохранения
os.makedirs(MODEL_DIR, exist_ok=True)
vectorizer_path = os.path.join(MODEL_DIR, "tfidf_vectorizer.joblib")

train_descriptions_for_fit = train_part[COL_DESCRIPTION].fillna("").tolist()

# 2. Проверка, существует ли уже обученный векторайзер (опционально)
if os.path.exists(vectorizer_path):
    print(f"Загрузка существующего TF-IDF векторайзера из {vectorizer_path}")
    vectorizer = joblib.load(vectorizer_path)
else:
    print("Обучение нового TF-IDF векторайзера на описаниях из train_part...")
    vectorizer = TfidfVectorizer(
        max_features=TFIDF_MAX_FEATURES,
        min_df=TFIDF_MIN_DF,
        max_df=TFIDF_MAX_DF,
        ngram_range=TFIDF_NGRAM_RANGE,
    )
    # Обучаем векторайзер ТОЛЬКО на описаниях из train_part
    vectorizer.fit(train_descriptions_for_fit)
    # Сохраняем обученный векторайзер (опционально)
    joblib.dump(vectorizer, vectorizer_path)
    print(f"TF-IDF векторайзер сохранён в {vectorizer_path}")

# 3. Трансформация описаний для всех датафреймов (train_part, val_part, test_full)
# Подготовим описания, заменив NaN на пустую строку
descriptions_train = train_part[COL_DESCRIPTION].fillna("").tolist()
descriptions_val = val_part[COL_DESCRIPTION].fillna("").tolist()
descriptions_test = test_full[COL_DESCRIPTION].fillna("").tolist()

# 4. Применение обученного векторайзера к описаниям из каждого датафрейма
print("Применение TF-IDF векторайзера к описаниям train_part...")
X_tfidf_train_sparse = vectorizer.transform(descriptions_train)
print("Применение TF-IDF векторайзера к описаниям val_part...")
X_tfidf_val_sparse = vectorizer.transform(descriptions_val)
print("Применение TF-IDF векторайзера к описаниям test_full...")
X_tfidf_test_sparse = vectorizer.transform(descriptions_test)

# 5. Преобразование разреженных матриц TF-IDF в pandas DataFrame
# Получаем имена признаков
tfidf_feature_names = [f"tfidf_{i}" for i in range(X_tfidf_train_sparse.shape[1])]

print(f"Создание DataFrame для TF-IDF признаков train_part (shape: {X_tfidf_train_sparse.shape})...")
train_tfidf_df = pd.DataFrame(X_tfidf_train_sparse.toarray(), columns=tfidf_feature_names, index=train_part.index)

print(f"Создание DataFrame для TF-IDF признаков val_part (shape: {X_tfidf_val_sparse.shape})...")
val_tfidf_df = pd.DataFrame(X_tfidf_val_sparse.toarray(), columns=tfidf_feature_names, index=val_part.index)

print(f"Создание DataFrame для TF-IDF признаков test_full (shape: {X_tfidf_test_sparse.shape})...")
test_tfidf_df = pd.DataFrame(X_tfidf_test_sparse.toarray(), columns=tfidf_feature_names, index=test_full.index)

# 6. Объединение TF-IDF признаков с основными датафреймами
print("Добавление TF-IDF признаков к train_part...")
train_part = pd.concat([train_part.reset_index(drop=True), train_tfidf_df.reset_index(drop=True)], axis=1)

print("Добавление TF-IDF признаков к val_part...")
val_part = pd.concat([val_part.reset_index(drop=True), val_tfidf_df.reset_index(drop=True)], axis=1)

print("Добавление TF-IDF признаков к test_full...")
test_full = pd.concat([test_full.reset_index(drop=True), test_tfidf_df.reset_index(drop=True)], axis=1)

print(f"Добавлено {len(tfidf_feature_names)} TF-IDF признаков к каждому датафрейму.")

# --- Вычисление признака количества жанров (book_genres_count) ---
# (Этот блок должен идти до create_aggregates, если genre_names используется в агрегатах или других признаках)
print("Вычисление признака количества жанров...")

def count_genres(genre_list):
    if isinstance(genre_list, list):
        return len(genre_list)
    else:
        return 0

train_part[F_BOOK_GENRES_COUNT] = train_part['genre_names'].apply(count_genres)
val_part[F_BOOK_GENRES_COUNT] = val_part['genre_names'].apply(count_genres)
test_full[F_BOOK_GENRES_COUNT] = test_full['genre_names'].apply(count_genres)

print("Признак количества жанров добавлен.")

# --- Вычисление агрегатов ---
print("Вычисление агрегированных признаков...")

# Вычисляем агрегаты на train_part и применяем к train_part и val_part
train_part = create_aggregates(train_part, train_part)
val_part = create_aggregates(val_part, train_part) # Используем агрегаты из train_part для val_part

```

```python
# -*- coding: utf-8 -*-

exclude_cols = ['user_id', 'book_id', 'rating', 'timestamp', 'has_read', 'description', 'title', 'author_name', 'genre_names', 'genre_names_str']
feature_cols = [col for col in train_part.columns if col not in exclude_cols]

X_train = train_part[feature_cols]
y_train_orig = train_part['rating']
X_val = val_part[feature_cols]
y_val_orig = val_part['rating']

import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error
from lightgbm import LGBMRegressor, early_stopping, log_evaluation
from catboost import CatBoostRegressor
from google.colab import files
import joblib # Для сохранения моделей (опционально)

# --- MODEL PARAMETERS ---
RANDOM_STATE = 42 # Убедитесь, что константа определена

LGB_PARAMS = {
    "objective": "rmse",
    "metric": "rmse",
    "n_estimators": 2000,
    "learning_rate": 0.01,
    "feature_fraction": 0.8,
    "bagging_fraction": 0.8,
    "bagging_freq": 1,
    "lambda_l1": 0.1,
    "lambda_l2": 0.1,
    "num_leaves": 31,
    "verbose": -1,
    # "n_jobs": -1,  # Не использовать с GPU
    "random_state": RANDOM_STATE,
    "boosting_type": "gbdt",
    "device": "gpu", # <-- Включить GPU
    "gpu_platform_id": 0,
    "gpu_device_id": 0,
}

LGB_FIT_PARAMS = {
    "eval_metric": "rmse",
    "callbacks": [
        early_stopping(stopping_rounds=50, verbose=True),
        log_evaluation(period=50)
    ],
}

CATBOOST_PARAMS = {
    'iterations': 1000,
    'learning_rate': 0.01,
    'depth': 5,
    'l2_leaf_reg': 0.1,
    'random_seed': RANDOM_STATE,
    'eval_metric': 'RMSE',
    'verbose': 50,
    'task_type': 'GPU',
    'devices': '0',
    'early_stopping_rounds': 50,
    'loss_function': 'RMSE'
}

# --- Функция для вычисления Score ---
def calculate_score(y_true, y_pred, target_range=10):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    normalized_rmse = rmse / target_range
    normalized_mae = mae / target_range
    score = 1 - (0.5 * normalized_rmse + 0.5 * normalized_mae)
    return score, rmse, mae

# --- Обучение LightGBM ---
print("--- Обучение LightGBM ---")
model_lgb = LGBMRegressor(**LGB_PARAMS)
model_lgb.fit(
    X_train,
    y_train_orig, # Используем оригинальный таргет
    eval_set=[(X_val, y_val_orig)], # Используем валидационную выборку
    **LGB_FIT_PARAMS
)
print("Обучение LightGBM завершено.")

# --- Обучение CatBoost ---
print("\n--- Обучение CatBoost ---")
model_cat = CatBoostRegressor(**CATBOOST_PARAMS)
model_cat.fit(
    X_train, y_train_orig, # Используем оригинальный таргет
    eval_set=[(X_val, y_val_orig)],
)
print("Обучение CatBoost завершено.")

# --- Предсказания на валидации ---
print("\n--- Предсказания на валидации ---")
y_val_pred_lgb = model_lgb.predict(X_val)
y_val_pred_cat = model_cat.predict(X_val)

# --- Оценка качества каждой модели отдельно ---
score_lgb, rmse_lgb, mae_lgb = calculate_score(y_val_orig, y_val_pred_lgb)
score_cat, rmse_cat, mae_cat = calculate_score(y_val_orig, y_val_pred_cat)

print("--- Результаты на валидации (LightGBM) ---")
print(f"Score: {score_lgb:.6f}, RMSE: {rmse_lgb:.6f}, MAE: {mae_lgb:.6f}")
print("--- Результаты на валидации (CatBoost) ---")
print(f"Score: {score_cat:.6f}, RMSE: {rmse_cat:.6f}, MAE: {mae_cat:.6f}")

# --- Простое усреднение предсказаний (ансамбль) на валидации ---
# Ограничиваем предсказания каждой модели [0, 10] перед усреднением
y_val_pred_lgb_clipped = np.clip(y_val_pred_lgb, 0, 10)
y_val_pred_cat_clipped = np.clip(y_val_pred_cat, 0, 10)

# Взвешенное или простое усреднение
weight_lgb = 0.5
weight_cat = 0.5


y_val_pred_ensemble = (weight_lgb * y_val_pred_lgb_clipped + weight_cat * y_val_pred_cat_clipped) / (weight_lgb + weight_cat)

score_ens, rmse_ens, mae_ens = calculate_score(y_val_orig, y_val_pred_ensemble)

print("--- Результаты на валидации (Ансамбль LGB + Cat) ---")
print(f"Score: {score_ens:.6f}, RMSE: {rmse_ens:.6f}, MAE: {mae_ens:.6f}")

# --- Предсказания для теста с помощью ансамбля ---
print("\n--- Предсказания для теста ---")
X_test = test_full[feature_cols]
y_test_pred_lgb = model_lgb.predict(X_test)
y_test_pred_cat = model_cat.predict(X_test)

# Ограничиваем предсказания [0, 10]
y_test_pred_lgb_clipped = np.clip(y_test_pred_lgb, 0, 10)
y_test_pred_cat_clipped = np.clip(y_test_pred_cat, 0, 10)

# Усреднение предсказаний для теста
y_test_pred_ensemble = (weight_lgb * y_test_pred_lgb_clipped + weight_cat * y_test_pred_cat_clipped) / (weight_lgb + weight_cat)


```
